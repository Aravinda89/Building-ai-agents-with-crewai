# 🌀 Memory that powers flows

In this second part of the course, we are going to build a sophisticated content creation system that uses CrewAI Flows to intelligently route different types of content to specialized agent teams/crews.

Here's the flow:
1. **Pass in a URL** from any website
2. **Determine the content type you want produced** (blog, newsletter, or LinkedIn post)
3. **The @router routes to specialized crews** based on the content type
4. **Content is generated** with the right format and tone

Think of it as having three different editorial teams, and a smart manager who assigns work to the right team!

Before we get started let go over why we need Flows for this use case. Single research crews work great for simple tasks, but real-world AI automation requires more sophisticated orchestration. **CrewAI Flows** provides this through an event-driven orchestration framework.

### 🧠 Define the memory for the Flow

- The state model, or memory, is like a shared notebook that all agents can read and write to. It persists data throughout the entire workflow execution.
- In CrewAI, it is defined as a Pydantic class.

```python
from typing import Dict, Any
from pydantic import BaseModel

# CODE: Define the state model and add its fields
class ContentState(BaseModel):
  """
  State model that tracks information throughout the content creation workflow.

  Think of this as a form that gets filled out as the workflow progresses:
  - Start: Only URL is filled
  - After routing: Content type is determined
  - After processing: Final content is ready
  - Throughout: Metadata can be added by any step
  """
  url: str = "" # CODE
  content_type: str = ""  
  final_content: str = ""  
  metadata: Dict[str, Any] = {} # CODE
```

**Why use state?**
- 💾 **Data Persistence**: Information flows between different workflow steps
- ✅ **Type Safety**: Pydantic ensures data integrity
- 🔄 **Flexibility**: Easy to add new fields as workflow evolves
- 🐞 **Debugging**: Clear view of what data exists at each step
