# 👋 Meet our first CrewAI Agent

What is an AI Agent? It can be thought of as an AI component with a specific job. It has:
- 🎭 a Role
- 🎯 a Goal 
- 📖 a Backstory
  
These elements define how it thinks and communicates. It also has a large language model attached:

- 🤖 the LLM

Which gives it the reasoning power to complete its tasks.

Here, we'll build a skilled autonomous Research Analyst Agent that can analyze web content and provide professional insights using a tool for web search.

### 🤖 Accessing an LLM

- We'll use a pre-configured OpenAI model throughout the course.

**❗️ Note: These will be hidden cells in the next exercises, as the ones below. Make sure to run them ❗️**

```python
from crewai import LLM

AGENT_LLM = LLM(model="openai/gpt-4o-mini")
```

### 🧰 Adding a tool
- Let's import the DuckDuckGo Search tool that our agent will use to search the web for content.

```python
!pip install -U --q ddgs duckduckgo-search langchain-community
```

- To use the DuckDuckGo tool, we'll need to wrap in the CrewAI tool definition.
- CrewAI has a function that does just that: We simply wrap the tool in a CrewAI **`@tool`** decorator and it handles converting it to a compatible CrewAI tool.

```python
from langchain_community.tools import DuckDuckGoSearchRun
from crewai.tools import tool

search_tool_instance = DuckDuckGoSearchRun()

# CODE: Wrap the search tool with the CrewAI @tool decorator
@tool("DuckDuckGo Search")
def search_tool(query: str) -> str:
    """A tool for searching the web using DuckDuckGo."""
    return search_tool_instance.run(query)
```

### 🧑‍💻 Running our first Agent

- We can now create a **Research agent** with role, goal, and backstory.
- We use the `AGENT_LLM` defined earlier to attach a model for the agent to use.

```python
from crewai import Agent

# CODE: Define the agent
research_agent = Agent(
  role="Senior Research Analyst",
  goal="Analyze the content of the provided website and extract key insights.",
  backstory="""You are a world-class research analyst with a knack for identifying
  the most important trends and data points from any text. You have a keen eye for detail
  and are known for your concise and insightful summaries.""",
  verbose=False,
  llm=AGENT_LLM,
  tools=[search_tool] # CODE: pass the wrapped tool
)

print("✅ Research Agent Created!")
print(f"Role: {research_agent.role}")
print(f"Goal: {research_agent.goal}")
```
